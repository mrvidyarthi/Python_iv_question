```python
def sum_digit(number: int) -> int:
    """
    This function is find the sum of given digit.
    Args:
        This function can accept only the integer number.
    Return:
        This funciton can return the sum of digit.
    Raise:
        TypeError: When the user are not input the integer value
        ValueError: When the user are enter less than aur equal 0
    """
    if not isinstance(number,int):
        raise TypeError("Inuput must be an integer.")
    if number <= 0:
        raise ValueError("Value are not less than and equal 0.")
    total = 0
    while number > 0:
        total = total + (number%10)
        number = number//10
    
    return total

if __name__ == "__main__":
    try: 
        user = input("Enter your digit : ")
        if not user.isdigit():
            raise TypeError("Input only the digit.")
        num = int(user)
        result = sum_digit(num)

        print(f"Sum of your digit: {result}")
    except (TypeError,ValueError,Exception) as error:
        print(error)
```
