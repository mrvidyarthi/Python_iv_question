```python
def largest_number(list1: list) -> list:
    """
    This function is find the largest number in given list
    Args:
        This function can take only the list
    Returns:
        This function is return None if list's number are same otherwise return largest number 
    """
    if not isinstance(list1, list):
        raise TypeError("Input must be one argunment.")
    
    if not list1:
        raise ValueError("Input must be one argument")
    larg_num = list1[0]
    all_same = True
    for item in list1:
        if not isinstance(item,(int,float)):
            raise TypeError("Input must be int or float.")
        if larg_num < item:
            larg_num = item
        if larg_num != item:
            all_same = False
    if all_same:
        return None
    else :
        return larg_num
if __name__ == "__main__":
    try : 
        user = input("Enter the number in a list seperated by space int or float only: ")
        
        list1 = [float(i) for i in user.split()]
        
        result = largest_number(list1)
        print(f"Largsert number in {list1} is : {result}")

    except (TypeError,ValueError) as e :
        print(e)
    except Exception as e :
        print(e)
    finally :
        print("Code are done sucssefully")
```
