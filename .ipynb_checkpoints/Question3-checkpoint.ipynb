```python
def is_prime(num : int) ->bool:
    """
    This function is wether that input number is Prime number or not.
    Args:
        This function is accept int type inpur.
    Returns:
        This funciton can return True if enter number is prime otherwise return False.
    """
    # Check the input in integer or not
    if not isinstance(num, int):
        raise TypeError("Enter must be integer.")

    # Logic for checking input is prime or not
    for i in range(2,(int(num**0.5)+1)):
        if num % i == 0:
            return False
            break
    else:
        return True

if __name__ == "__main__":
    try:
        user = input("Enter a number : ")
        if not user.strip().isdigit():
            raise TypeError("Input only number.")
        number = int(user)
        result = is_prime(number)
        if result:
            print(f"{number} is a prime number.")
        else:
            print(f"{number} is not a prime number.")
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print("Something went wrong ",error)
```
