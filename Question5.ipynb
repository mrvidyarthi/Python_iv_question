```python
def factorial(num : int) -> int:
    if not isinstance(num, int):
        raise TypeError("Input must be an integer.")
    if isinstance(num, bool):
        raise TypeError("Boolean are not allowed in this function.")
    fact = 1
    for i in range(1,num+1):
        fact = fact * i

    return fact
```

```python
def factorial(num : int) -> int:
    """
    This function is find the factorial of given input
    Args:
        This fucntion is accept integer value.
    Returns:
        This function is return the Factorial of given number.
    TypeError:
        If you are not insert the integer type value in this function.
        If you are insert the Boolean value.
    """
    
    if not isinstance(num, int):
        raise TypeError("Input must be an integer.")
        
    if isinstance(num, bool):
        raise TypeError("Boolean are not allowed in this function.")
        
    fact = 1
    for i in range(1,num+1):
        fact = fact * i

    return fact

if __name__ == "__main__":
    try : 
        user = input("Enter any number: ")
        if not user.strip().isdigit():
            raise ValueError("Input only the ineger number.")
        number = int(user)
        # calling funciton 
        result = factorial(number)
        print(f"Factorial of {number} is : {result}")
    except (TypeError,ValueError,Exception) as e :
        print(e)
        
```
