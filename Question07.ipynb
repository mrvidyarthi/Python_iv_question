```python
def vowels_count(string: str) -> int:
    """
    This function can count the number of vowels in given string.
    Args: 
        This function can take the string.
    Returns: 
        This function is return the number of vairables in given string.
    """
    vowels = "aeiou"
    count = 0
    if not isinstance(string, str):
        raise TypeError("Input must be string.")
    if string.isdigit():
        raise TypeError("Integer are not alowed. Please Enter the string.")
    
    for char in string.lower():
        if char in vowels:
            count += 1
    return count
if __name__ == "__main__" :
    try:
        string = input("Enter your string : ")
        result = vowels_count(string)
        print(f"""Total Vowels in "{string}" : {result}""")
    except (TypeError,ValueError,Exception) as error :
        print(error)
```
