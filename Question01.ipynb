```python

def reversed(string : str) -> str:
    """
    This function can be convert the reverse version of ginven string.
    Args : 
        This function is accept the string.
    Return: 
        This function can return the reverse version of string.
        
    """
    # Check the given string type is string type
    if not isinstance(string , str ):
        raise TypeError("Input must be string")

    # Logic for the revrse of string
    rev_str = ""
    for char in string:
        rev_str = char + rev_str
    return rev_str

if __name__ == "__main__":
    try:
        # Take input from the user
        user_input = input("Enter a string: ")
        # Calling the function
        if user_input.strip().isdigit():
            raise ValueError("Numeric value is not allowed.")
        result = reversed(user_input)
        # print the function 
        print(f"The reverse version of {user_input} : is {result}")
    except TypeError as error:
        print(error)
    except Exception as error:
        print(error)
```
