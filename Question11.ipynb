```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

```

```python
def factorial(n: int) -> int:
    """
    This function can find the factorial of given number using recursion.
    Args:
        This funciton is accerp the integer number
    Return:
        This function is return the factorial of given number.
    Raise:
        TypeError: If input are not integer.
        ValueError: If input negative value.
    """
    if isinstance(n,float):
        raise TypeError("The factorial of decimal are not possible.")
    if not isinstance(n, int):
        raise TypeError("Enter only the integer.")
    if n < 0 :
        raise ValueError("Factorial of negative number are not allowed.")
    
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

if __name__ == "__main__" :
    try : 
        user = input("Enter your number: ")
        if not user.isdigit():
            raise TypeError("Input only interger number.")
        user = int(user)
        result = factorial(user)
        print(f"Factorial: {result}")
    except (TypeError,ValueError,Exception) as e :
        print(e)
    finally:
        print("Code run successfully")
```
