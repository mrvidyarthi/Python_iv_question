```python
def fact(n):
    if n == 0 or n == 1:
        return n
    return n*fact(n-1)
```

```python
fact(5)
```

```python
def factorial(number : int) -> int:
    """
    This funciton calculate the factorial of given number.
    Args:
        number(int): This function accept only integer number.
    Returns:
        factorial(int): Return the factorial of given number.
    Raise:
        TypeError: If input are not integer.
        ValueError: If input are negative number.
    """
    if not isinstance(number,int):
        raise TypeError("Input must be an integer.")
    if number < 0:
        raise ValueError("Negative number are not allowed")
    if number == 0 or number == 1:
        return number 
    return number * factorial(number-1)
if __name__ == "__main__" :
    user = 6
    try: 
        result = factorial(user)
        print(f"Factorial of {user} is: {result}")
    except (TypeError, ValueError, Exception) as error:
        print(error)
```
