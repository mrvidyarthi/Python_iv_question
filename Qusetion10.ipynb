```python
def remove_duplicates(list1: list) -> list:
    """
    This function can remove the all duplicates value from the given list.
    Args:
        This function can take only the integer type list
    Returen:
        This funciton can return the new list in which removed all duplicates.
    Raise:
        TypeError: When the user are not insert the list type input.
        
    """
    if not isinstance(list1, list) :
        raise ValueError("Input only the list")
        
    new_list = []

    for item in list1:
        if item not in new_list:
            new_list.append(item)
    
    return new_list

if __name__ == "__main__" :
    try:
        user = input("Enter the number in a list seperated by space int or float only: ")
        
        lst = [int(i) for i in user.split()]

        result = remove_duplicates(lst)
        print("New list after remove duplicates.",result)
    except (TypeError,Exception) as e :
        print(e)
```
