```python
def anagram_checker(s1: str, s2: str) -> bool:
    
    """
    This function check that input two strings are anagram or not.
    Args:
        s1(str): Accept first string as a string.
        s2(str): Accept second string as a string.
    returns:
        Anagram(bool): Return True if anagram otherwise return False
    """
    
    if not isinstance(s1,str):
        raise TypeError("Input must be string")
    if not isinstance(s2,str):
        raise TypeError("Input must be string")
    
    s1 = s1.lower().replace(" ","")
    s2 = s2.lower().replace(" ","")
    
    anagram = True
    if len(s1) != len(s2):
        anagram = False
    else:
        for ch in s1:
            if s1.count(ch) != s2.count(ch):
                anagram = False
                
    return anagram

if __name__ == "__main__" :
    string1 = input("Enter your first string: ")
    string2 = input("Enter your second string: ")
    try:
        result = anagram_checker(string1,string2)
        if result:
            print(f"{string1} and {string2} both are anagram.")
        else:
            print(f"{string1} and {string2} both are not anagram.")
            
    except (TypeError,Exception) as error:
        print(error)
```
