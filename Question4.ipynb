```python
def is_prime(num: int) -> bool:
    """
    This funcition can find the input number is prime or not.
    Args :
        This function can take int number.
    Returns:
        This function is return True if number is prime otherwise return False
    """
    # Wether given number is int or not.
    if not isinstance(num,int):
        raise TypeError("Input must be an integer.")
    if num < 2:
        raise ValueError("Prime number is always more then 1.")
    # Logic buiding for wether that given nmber is prime or not
    for i in range(2,int(num**0.5)+1):
        if num%i == 0:
            return False
            break
    else :
        return True


if __name__ == "__main__" :
    try:
        lower_limit = input("Enter youe lower limit where you want to start range: ")
        upper_limit = input("Enter youe upper limit where you want to end range: ")
        if not lower_limit.strip().isnumeric() or not upper_limit.strip().isnumeric():
            raise TypeError("Please entern only the numeric value.")
        lower_limit = int(lower_limit)
        upper_limit = int(upper_limit)
    
        for i in range(lower_limit,upper_limit+1):
            if is_prime(i):
                print(f"{i} is prime number")
    except (TypeError,ValueError,Exception) as e:
        print(e)
```
