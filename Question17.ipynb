```python
def even_list(num:int) -> list:
    """This fucntion find the even number from 2 to enter number

    Args:
        num (int): Accept only number

    Returns:
        list: Even number item in a list
    """
    
    if not isinstance(num,int) :
        raise TypeError("Enter only integer number.")
    if num<2:
        raise ValueError("Even number is not less then 2")
    lst = []

    for i in range(2,num+1,2):
        lst.append(i)
    
    return lst

if __name__ == "__main__":
    n = int(input("Enter the last number: "))
    try:
        result = even_list(n)

        print(f"List of envn number: {result}")
    except (TypeError,ValueError, Exception) as e :
        print(e)
```
