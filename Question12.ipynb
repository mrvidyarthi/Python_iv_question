```python
def char_count(string : str) -> dict:
    """This function can count the each character in your string.

    Args:
        string (str): This function can accept only the string.

    Returns:
        dict: funciton can return the dictionary.
    """
    
    ch_count = {}
    
    
    for ch in string.lower().replace(" ",""):
        
        if ch not in ch_count:
            ch_count[ch] = 1
        else:
            ch_count[ch] += 1
            
    return ch_count

if __name__ == "__main__" :
    user = input("Enter your string/lines: ")
    try: 
        result = char_count(user)
        print(f"You are input: {user}\nTotal Occurrences of each element {result}")
    except (TypeError,Exception) as e:
        print(e)
```
