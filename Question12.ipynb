```python
def character_count(string : str ) -> str:
    """
    This function is count the character in your string.
    Args:
        This function can take only the string.
    Returns:
        This function can return the count of character in your string
    """
    if not isinstance(string,str):
        raise TypeError("Please input only the string.")
    char_count = {}
    
    for char in string.lower():
        if char in char_count:
            char_count[char] = char_count[char] + 1
    
        else:
            char_count[char] = 1
    
    return char_count
if __name__ == "__main__" :
    try:
        user = input("Enter your string.")
        result = character_count(user)
        print("")
```
