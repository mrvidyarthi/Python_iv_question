```python
def second_largest_num(lst: list) -> (float,bool):
    """
    This function can find the second largest number from a list which list is store numeric value.
    Args:
        This function is accept list in which int or float only
    Returns:
        None: If All numbers are equal the return None
        Second lrgest number: return second largest number 
        
    """
    if not isinstance(lst,list):
        raise TypeError("Input must be list")
    if len(lst) < 2:
        raise ValueError("Please enter the number of item in list must be grater then 1")
    all_same = True 
    larg_num = lst[0]
    secd_larg = None
    for num in lst:
        if larg_num != num:
            all_same = False 
        if larg_num < num :
            secd_larg = larg_num
            larg_num = num
        elif num != larg_num and (secd_larg is None or num > secd_larg):
            secd_larg = num

    if all_same:
        return None
    else:
        return secd_larg
if __name__ == "__main__" : 
    try:
        lst = input("Enter the number in list separeted by space int or float : ")
        list1 = [float(i) for i in lst.split()]
 
        result = second_largest_num(list1)

        print(f"Second largest number in your list: {result}")
    except (ValueError,TypeError,Exception) as error:
        print(error)
```
