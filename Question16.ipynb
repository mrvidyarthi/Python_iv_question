```python
def power_generater(x:int, y:int) -> int:
    """This function calculate the power of any number

    Args:
        x (int): For base number
        y (int): Power of base number

    Returns:
        int: Power of x
    """
    power = 1

    for i in range(y):
        power = power * x
    return power
        
```

```python
print(power_generater(2,10))
```

```python
class Power_Genterate:
    def __init__(self):
        pass
    def power_of_any_number(self,x:int, y:int) -> int:
        """This function calculate the power of any number

        Args:
            x (int): For base number
            y (int): Power of base number

        Returns:
            int: Power of x
        """
        
        if not isinstance(x,int) or not isinstance(y,int):
            raise TypeError("Input must be an integer.")
        power = 1
        
        if y<0:
            raise ValueError("Power of any number is not be a negative.")
        
        for i in range(y):
            power = power * x
        return power
    
if __name__ == "__main__":
    try:
        base = int(input("Enter your base number: "))
        power_num = int(input(f"Enter the power of {base}: "))
        
        Object_of_Power_Generator = Power_Genterate()
        result = Object_of_Power_Generator.power_of_any_number(base,power_num)
        print(f"{base}^{power_num} = {result}")
       
    except (TypeError,ValueError) as e :
        print(e)
            
```
