```python
def check_palindrome(string: str) -> bool:
    """
    This function is check that input is palindrome or not
    Args:
        This function is take string
    Returns:
        This fuction is Return True if given string is palindrome other wise return False.
    TypeError:
        If input is not a string.
    """
    if not isinstance(string,str):
        raise TypeError("Input must be string.")
    if not string:
        raise ValueError("Input must be one argument.")
        
    string = string.lower()
    rev_str = ""
    for i in string:
        rev_str = i + rev_str
    if string == rev_str:
        return True
    else:
        return False

if __name__ == "__main__":
    try:
        string = input("Enter a your string: ")
        result = check_palindrome(string)

        if result:
            print(f"Yes {string} is a palindorme string")
        else:
            print(f"No {string} is not a palindorme string")
    except (TypeError,ValueError,Exception) as error:
        print(error)
```
